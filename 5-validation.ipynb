```python
def plot_matrix(mat,title,cmap,vmax,vmin=0):
    plt.figure()
    plt.imshow(mat,interpolation='nearest',cmap=cmap,vmax=vmax,vmin=vmin)
    plt.colorbar()
    plt.title(title)
    plt.show()
```

```python
plot_matrix(matrix_all[1,:],"yes","Reds",vmax=1,vmin=0)
plot_matrix(matrix1,"cooler","Reds",vmax=1,vmin=0)
```
