```python
chrom_list = df_data['chrom1'].unique().tolist()
outputdir = "/shareb/mliu/evaluate_impute/data/real_sc_hic/Ramani/paris"
```

```python
header_lines_cell= [
"## pairs format v1.0",
"#sorted: chr1-chr2-pos1-pos2",
"#shape: upper triangle",
"#chromosome: chr1 249250621",
"#chromosome: chr2 243199373",
"#chromosome: chr3 198022430",
"#chromosome: chr4 191154276",
"#chromosome: chr5 180915260",
"#chromosome: chr6 171115067",
"#chromosome: chr7 159138663",
"#chromosome: chr8 146364022",
"#chromosome: chr9 141213431",
"#chromosome: chr10 135534747",
"#chromosome: chr11 135006516",
"#chromosome: chr12 133851895",
"#chromosome: chr13 115169878",
"#chromosome: chr14 107349540",
"#chromosome: chr15 102531392",
"#chromosome: chr16 90354753",
"#chromosome: chr17 81195210",
"#chromosome: chr18 78077248",
"#chromosome: chr19 59128983",
"#chromosome: chr20 63025520",
"#chromosome: chr21 48129895",
"#chromosome: chr22 51304566",
"#chromosome: chrX 155270560",
"#columns: chr1 pos1 chr2 pos2"
    ]
```

```python
for cell_id in range(620):
    df_data_cell = df_data[df_data['cell_id']==cell_id]
    
    df_data_cell.sort_values(by=['chrom1','pos1','chrom2','pos2'],inplace=True)
    df_data_cell.index = range(len(df_data_cell))
    df_data_cell = df_data_cell[['chrom1','pos1','chrom2','pos2']]
    #转成pairs
    filename = f"{outputdir}/cell{cell_id}.pairs.gz"
    csv_buffer = StringIO()
    for line in header_lines_cell:
        csv_buffer.write(line + '\n')
    df_data_cell.to_csv(csv_buffer, index=False, header=False,sep="\t")
    with gzip.open(filename, 'wt') as f:
            f.write(csv_buffer.getvalue())
    
```
